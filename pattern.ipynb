```python
def generate_hollow_square(n):
    pattern=[]
    for i in range(n):
        if i==0 or i==n-1:
            pattern.append(n*'*')
        else:
            # pattern.append('*'+(n-2*' ')+'*')
            pattern.append('*' + ' ' * (n - 2) + '*')
    return pattern

print(generate_hollow_square(5))
```

Triangular Pattern 

```python
def generate_triangular_pattern(n):
    pattern=[]
    for i in range(1,n+1):
        stars = 2*(n-i) - 1
        spaces = (2*n -1)-stars
        pattern.append(' '*(spaces//2) + '*'*stars+' '*(spaces//2))
        print(' '*(spaces//2) + '*'*stars+' '*(spaces//2))
    return pattern
# generate_triangular_pattern(int(input("Enter the size of the square: ")))

def generate_diamond_pattern(n):
    rows = 2*n -1
    for i in range(int(rows)):
        if i<n:
            stars = 2*i + 1
            spaces = (2*n -1)-stars
            print(' '*(spaces//2) + '*'*stars+' '*(spaces//2))
        else:
            stars = 2*(rows - i-1)+1
            spaces = (2*n -1)-stars
            print(' '*(spaces//2) + '*'*stars+' '*(spaces//2))
    
generate_diamond_pattern(int(input("Enter the size of the square: ")))

```

Floyds Triangle
1
2 3
4 5 6

```python
def generate_floyds_triangle(n):
    pattern=[]
    num = 0
    for i in range(n):
        val=''
        for j in range(i+1):
            num+=1
            if j==i:
                val+=str(num)
            else:
                val+=str(num)+' '
            
        pattern.append(val)
    return pattern

print(generate_floyds_triangle(5))

        
```
