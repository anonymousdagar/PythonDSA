Linear Search - Search in loop one by one

```python
def linear_search(arr, target):
    for index in range(len(arr)):
        if arr[index]==target:
            return index
    return -1
```

Binary Search - Applicable on sorted list/array

```python
def binary_search(arr,target):
    start,end=0,len(arr)-1
    while start<=end:
        middle = (start+end)//2
        if arr[middle]==target:
            return middle
        elif arr[middle]<target:
            start=middle+1
        else:
            end=middle-1

    return -1
print(binary_search([2,3,4,5,6,33,54,645,654,666,754,876,2345],26))
```

Bubble Sorting 

```python
def bubble_sort(lst):
    # Your code goes here
    size = len(lst)
    correct_positions =0
    while correct_positions <size:
        for i in range(len(lst)-1-correct_positions):
            if lst[i]>lst[i+1]:
                lst[i],lst[i+1]=lst[i+1],lst[i]
        correct_positions+=1
    return lst
    
            
print(bubble_sort([6,5,4,3,2,1]))
```

Selection Sort

```python
def selection_sort(lst):
    # Your code goes here
    correct_position=0
    while correct_position<len(lst)-1:
        index,val=correct_position,lst[correct_position] 
        for i in range(correct_position,len(lst)):
            if lst[i]<val:
                index=i
                val=lst[i]
      
        lst[correct_position],lst[index]=lst[index],lst[correct_position]
        correct_position+=1
    return lst
print(selection_sort([6,5,4,3,2,23,4]))
```

Insertion Sort

```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        if arr[i]>=arr[i-1]:
            continue
        else:
            num = arr[i]
            j=i
            while arr[j-1]>num and j>0:
                arr[j]=arr[j-1]
                j-=1
            arr[j]=num
    return arr

print(insertion_sort([23,5,34,21,432,52,554,2,3,36,89]))
            
```

1. Count negative numbers in a sorted matrix

```python
def countNegatives(grid):
    c=0
    for lst in grid:
        size = len(lst)
        i,j=0,size-1
        while i<j:
            mid=(i+j)//2
            if lst[mid]>=0:
                i=mid+1
            else:
                j=mid-1
        if lst[i]<0:
            c+=size-i
        else:
            c+=size-i-1
    return c
print(countNegatives([[4, 3, 2, 1], [3, 2, 1, -1], [1, 1, -1, -2], [-1, -1, -2, -3]]))
```

Minimum in rotated sorted array

```python
def findMin(nums):
    i,j=0, len(nums)-1
    while i<=j:
        mid=(i+j)//2
        if nums[mid]>nums[j]:
            i=mid+1
        else:
            j=mid
    return nums[i]
print(findMin( [4, 5, 6, 7, 0, 1, 2] ))
```

Search in Rotated Sorted Array (Very Good One)

```python
def search(nums, target):
    i,j=0,len(nums)-1
    while i<=j:
        mid = (i+j)//2
        if nums[mid]==target:
            return mid
        # Left half sorted
        if nums[i]<=nums[mid]:
            if nums[i]<=target<nums[mid]:
                j=mid -1
            else:
                i=mid+1
        # right half sorted
        else:
            if nums[mid]<target<=nums[j]:
                i=mid+1
            else:
                j=mid-1
    return -1
```
