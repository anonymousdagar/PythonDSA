Multiple Inheritance

```python
# Base Class 1
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("SubClass must implement the method")

# Base Class 2
class Pet:
    def __init__(self,owner):
        self.owner = owner

# Derived Class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} says woof"


# Creating Object
dog = Dog("buddy","Krinsh")

print(dog.owner)
dog.speak()
```

```python
class BankAccount:
    account_Counter=1000
    def __init__(self,account_holder, initial_balance=0):
        self.account_holder=account_holder
        self.initial_balance=initial_balance
        self.account_number = BankAccount.account_Counter
        BankAccount.account_Counter+=1
        
    def balance(self):
        return self.initial_balance
    
    def deposit(self,amount):
        if amount>0:
            self.initial_balance+=amount
        else:
            raise ValueError("Deposit amount muse be positive")
    def withdraw(self,amount):
        if amount>0:
            if self.initial_balance>=amount:
                self.initial_balance-=amount
            else:
                raise ValueError("Insufficiant Balance")
        else:
            raise ValueError("Withdrawal amount must be positive")
            
    def transfer(self,account,amount):
        if not isinstance(account,BankAccount):
            raise TypeError("Transfer must be to another BankAccount")
        if amount>0:
            if self.initial_balance>=amount:
                self.initial_balance-=amount
                account.initial_balance+=amount
            else:
                raise ValueError("Insufficiant Funds")
        else:
            raise ValueError("Tranfer amount must be positive")
    
    def __repr__(self):
        return f"BankAccount({self.account_holder}, Balance:{self.initial_balance} ."

acc1=BankAccount("Pankaj",5500)
acc2=BankAccount("Priyanka",200)

# acc1.deposit(3000)
acc1.balance()

acc1.transfer(acc2,3000)
acc1.balance()
acc2.balance()

        
```
