Factorial

```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)

factorial(5)
```

Sun of N numbers

```python
def sum_of_naturalNumbers(n):
    if n==1:
        return 1
    return n + sum_of_naturalNumbers(n-1)
sum_of_naturalNumbers(10)
```

Number of Digits in a Number

```python
def count_digits(n):

    if n>=1 and n<=9:
        return 1
    elif n==0:
        return 1
    
    return count_digits(int(n/10))+1    

print(count_digits(1234))
```

Printing 1 to N numebrs into list

```python
def count_to_n(n):
    if n==1:
        return [1]
    return [n] +count_to_n(n-1)

```

```python
def minAllOneMultiple( k: int) -> int:
    n=1
    while n%k!=0:
        n=int(str(n)+'1')
    return len(str(n))
print(minAllOneMultiple(2))
```

Check if an array is sorted

```python
def check_sortedArray(arr,i=0):
    n=len(arr)-1
    if i==n:
        return True
    if arr[i]<=arr[i+1]:
        return check_sortedArray(arr,i+1)
    else:
        return False
    
check_sortedArray([2,6,9,88])
```

Find First Index of Element

```python
def find_first_index(arr, element):

    if len(arr)==0:
        return -1
    if arr[0]==element:
        return 0
    else:
        ans = find_first_index(arr[1:],element)
        if ans==-1:
            return ans
        else:
            return ans+1
print(find_first_index([1,2,3,4,4,5,6],2))    
```

Write Program of find all index of a given Element

```python
def find_indices(arr, element):
    def helper(index):
        if index==len(arr):
            return []
        if arr[index]==element:
            return [index]+helper(index+1)
        else:
            return helper(index+1)
    
    return helper(0)
    

    
print(find_indices([1,2,4,6,3,2,54],2))
    
```

```python
def merge(nums1,m,nums2,n):
    i,j,k=m-1,n-1,m+n-1
    while i>=0 and j>=0:
        if nums1[i]>nums2[j]:
            nums1[k]=nums1[i]
            i-=1
        else:
            nums1[k]=nums2[j]
            j-=1
        k-=1
    while j>=0:
        nums1[k]=nums2[j]
        k-=1
        j-=1
    return nums1
print(merge([1,2,3,0,0,0],3,[2,5,6],3))
            
```
