```python
def is_prime(num):
    prime=True
    for i in range(2,num):
        if num%i==0:
            prime = False
    return prime
print(is_prime(111))
```

```python
def hcf(m,n):
    start=2
    largest_Hcf =1
    if m<n:
        start=m
        while start>1:
            if m%start==0 and n%start==0:
                largest_Hcf=start
                break
            start-=1
    else:
        start = n
        while start>1:
            if m%start==0 and n%start==0:
                largest_Hcf=start
                break
            start-=1
    return largest_Hcf


string = 'hello there myself pankaj dagat'
newstr=''
for char in string:
    if char not in newstr:
        newstr+=char
print(newstr)
```

```python
def rangeFinder(digits):
    short,long=[],[]
    for i in range(len(digits)):
        if digits[i]!=0:
            short.append(digits[i])
        else:
            long.append(digits[i])
    return short+long
print(rangeFinder([0, 0, 1]))
```

```python
def intersection(num1,num2):
    return [i for i in num1 for j in num2 if i==j]
print(intersection([1,2,3],[4,2,6]))
```
