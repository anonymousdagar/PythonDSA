Read Whole File

```python
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

Reading line by line

```python
with open ('example.txt','r') as file:
    for line in file:
        print(line)
```

Write in File

```python
with open('example.txt','w') as file:
    file.write('Writer once mmndk')
    file.write("issue Resolved")
```

```python
lines=['First Line \n','Second Line \n']
with open('example.txt','a') as file:
     file.write('Appended using a code ')
     file.writelines(lines)
```
